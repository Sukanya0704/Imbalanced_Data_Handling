```python
import torch
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import warnings
warnings.filterwarnings('ignore')

# GPU Verification
print(f"PyTorch version: {torch.__version__}")
print(f"CUDA available: {torch.cuda.is_available()}")

if torch.cuda.is_available():
    device_name = torch.cuda.get_device_name(0)
    print(f"🎯 GPU detected: {device_name}")
    print(f"🔧 CUDA version: {torch.version.cuda}")
    print(f"💾 GPU Memory: {torch.cuda.get_device_properties(0).total_memory / 1024**3:.1f} GB")
    
    # Test GPU with a simple operation
    x = torch.randn(3, 3).cuda()
    print(f"✅ GPU test: Tensor successfully moved to GPU - {x.device}")
else:
    print("❌ WARNING: CUDA not available")

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import (precision_score, recall_score, f1_score,
                             balanced_accuracy_score, matthews_corrcoef,
                             roc_auc_score, average_precision_score,
                             confusion_matrix)

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from xgboost import XGBClassifier

from ctgan import CTGAN
import joblib

os.environ["JOBLIB_START_METHOD"] = "spawn"
```

```python
# GPU Memory Management
def setup_gpu():
    if torch.cuda.is_available():
        torch.cuda.empty_cache()
        torch.backends.cudnn.benchmark = True  # Optimize for fixed input sizes
        print("✅ GPU optimized and memory cleared")
        
        # Show initial memory usage
        print(f"📊 Initial GPU memory allocated: {torch.cuda.memory_allocated()/1024**3:.2f} GB")
        print(f"📊 Initial GPU memory cached: {torch.cuda.memory_reserved()/1024**3:.2f} GB")
        return True
    return False

# Setup GPU
gpu_available = setup_gpu()

if gpu_available:
    print("🚀 Training will use GPU (much faster!)")
    print("⏰ Expected training time: 15-30 minutes for 100 epochs")
else:
    print("⚠️ Training will use CPU (slower)")
```

```python
# Load your Kaggle Credit Card Fraud dataset
data = pd.read_csv("D:\\stc\\dataset\\creditcard.csv")

# Quick look
print("Dataset shape:", data.shape)
print("\nFirst 5 rows:")
print(data.head())
print("\nClass distribution:")
print(data['Class'].value_counts())
print(f"Fraud percentage: {data['Class'].mean()*100:.4f}%")

# Plot class imbalance
plt.figure(figsize=(10, 6))
sns.countplot(x='Class', data=data)
plt.title("Class Distribution (0 = Legit, 1 = Fraud)")
plt.show()
```

```python
# Split data
train, test = train_test_split(data, test_size=0.2, random_state=42, stratify=data['Class'])
train, val = train_test_split(train, test_size=0.2, random_state=42, stratify=train['Class'])

print(f"📊 Dataset splits:")
print(f"Train size: {len(train)} ({(len(train)/len(data)*100):.1f}%)")
print(f"Validation size: {len(val)} ({(len(val)/len(data)*100):.1f}%)")
print(f"Test size: {len(test)} ({(len(test)/len(data)*100):.1f}%)")
print(f"Frauds in train: {train['Class'].sum()} ({(train['Class'].mean()*100):.3f}%)")
print(f"Frauds in test: {test['Class'].sum()} ({(test['Class'].mean()*100):.3f}%)")
```

```python
# Scale only Time and Amount columns
scaler = StandardScaler()
cols_to_scale = ['Time', 'Amount']

train_scaled = train.copy()
val_scaled = val.copy()
test_scaled = test.copy()

train_scaled[cols_to_scale] = scaler.fit_transform(train[cols_to_scale])
val_scaled[cols_to_scale] = scaler.transform(val[cols_to_scale])
test_scaled[cols_to_scale] = scaler.transform(test[cols_to_scale])

print("✅ Data preprocessing completed!")
print(f"Columns scaled: {cols_to_scale}")
```

```python
# Diagnostic Cell - Run this first
print("🔍 Running GPU and CTGAN diagnostics...")

# Check GPU status
if torch.cuda.is_available():
    print(f"✅ GPU is available: {torch.cuda.get_device_name(0)}")
    print(f"📊 GPU memory allocated: {torch.cuda.memory_allocated()/1024**3:.2f} GB")
    print(f"📊 GPU memory reserved: {torch.cuda.memory_reserved()/1024**3:.2f} GB")
else:
    print("❌ GPU not available")

# Check data size
print(f"📊 Training data shape: {train_scaled.shape}")
print(f"📊 Data types: {train_scaled.dtypes[:5]}...")  # Show first 5 columns

# Check for NaN values
print(f"🔍 NaN values in data: {train_scaled.isna().sum().sum()}")

# Test a small CTGAN instance
print("\n🧪 Testing with tiny CTGAN instance...")
try:
    test_ctgan = CTGAN(epochs=2, verbose=True, cuda=True)
    # Test with small subset
    test_data = train_scaled.head(1000)
    test_ctgan.fit(test_data, discrete_columns=['Class'])
    print("✅ Small test successful! The issue is with full dataset.")
except Exception as e:
    print(f"❌ Small test failed: {e}")
```

```python
# CTGAN Training - ULTRA OPTIMIZED VERSION
print("🚀 Starting ultra-optimized CTGAN training...")

# Use MUCH smaller dataset - focus on quality over quantity
fraud_samples = train_scaled[train_scaled['Class'] == 1]
non_fraud_samples = train_scaled[train_scaled['Class'] == 0]

# Very small subset that will definitely work
fraud_sample_size = len(fraud_samples)  # Keep all frauds (394 samples)
non_fraud_sample_size = 10000  # Only 10K non-frauds

non_fraud_subset = non_fraud_samples.sample(n=non_fraud_sample_size, random_state=42)
train_small = pd.concat([fraud_samples, non_fraud_subset], ignore_index=True)

print(f"📊 Using very small training subset:")
print(f"   Fraud samples: {len(fraud_samples)}")
print(f"   Non-fraud samples: {len(non_fraud_subset)}")
print(f"   Total: {len(train_small)} samples")
print(f"   Memory usage: {train_small.memory_usage(deep=True).sum() / 1024**2:.1f} MB")

# Clear GPU memory aggressively
if torch.cuda.is_available():
    torch.cuda.empty_cache()
    import gc
    gc.collect()
    print("🧹 GPU memory aggressively cleared")

# Minimal CTGAN parameters
ctgan = CTGAN(
    epochs=50,           # Reasonable epochs
    batch_size=64,       # Very small batch size
    verbose=True,
    cuda=True,
    generator_dim=(64, 64),       # Very small networks
    discriminator_dim=(64, 64),   # Very small networks
    generator_lr=2e-4,
    discriminator_lr=2e-4,
    pac=1                        # Essential for small batch sizes
)

print("⏳ Training starting NOW...")
print("   You should see progress within 15 seconds!")

# Train on the small dataset
ctgan.fit(train_small, discrete_columns=['Class'])

# Save model
joblib.dump(ctgan, 'ctgan_gpu_trained.pkl')
print("✅ CTGAN training completed successfully!")
```

```python
# Generate synthetic fraud samples
print("🎯 Generating synthetic fraud samples...")

fraud_count = len(train_scaled[train_scaled['Class'] == 1])
non_fraud_count = len(train_scaled[train_scaled['Class'] == 0])
target_fraud_fraction = 0.05  # 5% fraud rate target

# Compute how many synthetic frauds we need
total_target = int(non_fraud_count / (1 - target_fraud_fraction))
synthetic_needed = total_target - (fraud_count + non_fraud_count)

print(f"📊 Current stats:")
print(f"   Original frauds: {fraud_count}")
print(f"   Original non-frauds: {non_fraud_count}")
print(f"   Target fraud ratio: {target_fraud_fraction*100}%")
print(f"   Synthetic frauds needed: {synthetic_needed}")

# Generate synthetic fraud samples in batches
print("🔄 Generating synthetic data in batches...")
batch_size = 1000
synthetic_fraud_list = []
generated_count = 0

while generated_count < synthetic_needed:
    remaining = synthetic_needed - generated_count
    current_batch_size = min(batch_size, remaining * 2)  # Generate extra
    
    print(f"   Generating batch of {current_batch_size} samples...")
    synthetic_batch = ctgan.sample(current_batch_size)
    synthetic_fraud_batch = synthetic_batch[synthetic_batch['Class'] == 1]
    
    synthetic_fraud_list.append(synthetic_fraud_batch)
    generated_count += len(synthetic_fraud_batch)
    
    print(f"   ✅ Got {len(synthetic_fraud_batch)} fraud samples (Total: {generated_count}/{synthetic_needed})")
    
    # Clear GPU memory between batches
    if torch.cuda.is_available():
        torch.cuda.empty_cache()

# Combine all batches
synthetic_fraud = pd.concat(synthetic_fraud_list, ignore_index=True).head(synthetic_needed)

print(f"🎉 Successfully generated {len(synthetic_fraud)} synthetic fraud samples")

# Combine with original train data
train_gan = pd.concat([train_scaled, synthetic_fraud], ignore_index=True)

print("\n📈 Final dataset comparison:")
print(f"   Original train: {len(train_scaled)} samples, fraud ratio: {(train_scaled['Class'].mean()*100):.3f}%")
print(f"   GAN-augmented: {len(train_gan)} samples, fraud ratio: {(train_gan['Class'].mean()*100):.2f}%")
print(f"   Added {len(synthetic_fraud)} synthetic fraud samples")
```

```python
# Prepare features and targets
print("📊 Preparing features for model training...")

X_train_orig = train_scaled.drop('Class', axis=1)
y_train_orig = train_scaled['Class']
X_train_gan = train_gan.drop('Class', axis=1)
y_train_gan = train_gan['Class']
X_test = test_scaled.drop('Class', axis=1)
y_test = test_scaled['Class']

print(f"Feature shapes:")
print(f"   X_train_orig: {X_train_orig.shape}")
print(f"   X_train_gan: {X_train_gan.shape}")
print(f"   X_test: {X_test.shape}")

# Define models for comparison
models = {
    "DecisionTree": DecisionTreeClassifier(random_state=42),
    "RandomForest": RandomForestClassifier(random_state=42, n_jobs=1),
    "LogisticRegression": LogisticRegression(max_iter=1000, random_state=42),
    "SVM": SVC(probability=True, random_state=42),
    "XGBoost": XGBClassifier(eval_metric='logloss', random_state=42, n_jobs=1)
}

results = []
print("✅ Feature preparation completed!")
```

```python
# Comprehensive evaluation function
def evaluate_model(model, X_test, y_test, name, cost_fn=1000, cost_fp=10):
    y_pred = model.predict(X_test)
    y_prob = model.predict_proba(X_test)[:, 1]
    
    precision = precision_score(y_test, y_pred, zero_division=0)
    recall = recall_score(y_test, y_pred, zero_division=0)
    f1 = f1_score(y_test, y_pred, zero_division=0)
    bal_acc = balanced_accuracy_score(y_test, y_pred)
    mcc = matthews_corrcoef(y_test, y_pred)
    roc = roc_auc_score(y_test, y_prob)
    pr = average_precision_score(y_test, y_prob)
    
    tn, fp, fn, tp = confusion_matrix(y_test, y_pred).ravel()
    baseline_cost = sum(y_test) * cost_fn
    model_cost = fn * cost_fn + fp * cost_fp
    saving = (baseline_cost - model_cost) / baseline_cost * 100 if baseline_cost > 0 else 0
    
    return {
        'Model': name,
        'Precision': precision,
        'Recall': recall,
        'F1': f1,
        'Balanced Accuracy': bal_acc,
        'MCC': mcc,
        'ROC AUC': roc,
        'PR AUC': pr,
        'Cost Saving (%)': saving,
        'TP': tp,
        'FP': fp,
        'TN': tn,
        'FN': fn
    }

print("✅ Evaluation function defined!")
```

```python
# Train and evaluate models
print("🤖 Training and evaluating models...")
print("="*60)

for name, clf in models.items():
    print(f"\n🔧 {name}")
    print("-" * 40)
    
    # Train on original data
    print("Training on original data...", end=" ")
    clf_orig = clf
    clf_orig.fit(X_train_orig, y_train_orig)
    res_orig = evaluate_model(clf_orig, X_test, y_test, f"{name} (Original)")
    results.append(res_orig)
    print(f"✅ F1: {res_orig['F1']:.4f}")
    
    # Print detailed scores for original data
    print(f"   Precision: {res_orig['Precision']:.4f}")
    print(f"   Recall: {res_orig['Recall']:.4f}")
    print(f"   Balanced Accuracy: {res_orig['Balanced Accuracy']:.4f}")
    print(f"   ROC AUC: {res_orig['ROC AUC']:.4f}")
    print(f"   Cost Saving: {res_orig['Cost Saving (%)']:.2f}%")
    print(f"   Confusion Matrix: TP={res_orig['TP']}, FP={res_orig['FP']}, TN={res_orig['TN']}, FN={res_orig['FN']}")

    # Train on GAN-augmented data
    print("Training on GAN-augmented data...", end=" ")
    clf_gan = clf.__class__(**clf.get_params())
    clf_gan.fit(X_train_gan, y_train_gan)
    res_gan = evaluate_model(clf_gan, X_test, y_test, f"{name} (CTGAN)")
    results.append(res_gan)
    print(f"✅ F1: {res_gan['F1']:.4f}")
    
    # Print detailed scores for GAN data
    print(f"   Precision: {res_gan['Precision']:.4f}")
    print(f"   Recall: {res_gan['Recall']:.4f}")
    print(f"   Balanced Accuracy: {res_gan['Balanced Accuracy']:.4f}")
    print(f"   ROC AUC: {res_gan['ROC AUC']:.4f}")
    print(f"   Cost Saving: {res_gan['Cost Saving (%)']:.2f}%")
    print(f"   Confusion Matrix: TP={res_gan['TP']}, FP={res_gan['FP']}, TN={res_gan['TN']}, FN={res_gan['FN']}")

print("\n🎉 All models trained and evaluated!")
```
